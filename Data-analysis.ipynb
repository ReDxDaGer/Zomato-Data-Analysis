# Pie Chart for the Countries

```python
Country_value = final_df.Country.value_counts().values
```

## This Pie chart is for the computation of percentage between top two countries that uses Zomato

```python
plt.pie(Country_value[:3],labels=country_count[:3],autopct='%1.1f%%')
plt.show()
```

**Obeservation: Zomato maximum record or transaction are from India After that USA and then UK (Hence Major Business is going on in India)**

```python
final_df.columns
```

```python
ratings = final_df.groupby(['Aggregate rating' , 'Rating color' , 'Rating text'])['Rating text'].size().reset_index(name='Rating Count')
```

```python
ratings
```

## Observations
- When the ratings is between 4.5 to 4.9 --> Excellent
- When the ratings is between 4.0 to 4.4 --> Very Good
- When the ratings is between 3.5 to 3.9 --> Good
- When the ratings is between 3.0 to 3.4 --> Average
- When the ratings is between 2.5 to 2.9 --> Average
- When the ratings is between 2.0 to 2.4 --> Poor

**Most of the people above didn't rated as we can see below in the BarPlot**

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12, 6)

sns.barplot(x='Aggregate rating', y='Rating Count',hue='Rating color' ,data=ratings , palette =['purple','red','orange','yellow','green'])
plt.xlabel('Rating')
plt.ylabel('Count')
plt.title('Distribution of Ratings')

# Show the plot
plt.show()
```

## Obesrvations: 
- Not Rated count is very high
- Maximum number of rating are between 2.5 to 3.5
